3. Load the CSV metadata

4. Simple data analysis (EDA)

```python
# ===== Basic statistics and structure =====

print("Shape:", df_attr.shape)
print("\nColumn types:")
print(df_attr.dtypes)

print("\nMissing values per column:")
print(df_attr.isna().sum())

print("\n----- Numerical Summary -----")
display(df_attr.describe(include="all"))
```

```python
# ===== Age distribution =====

# Age is stored as "###d" (days). Convert to integer days.
df_attr["Age_days"] = df_attr["Age"].str.replace("d", "").astype(int)
df_attr["Age_years"] = df_attr["Age_days"] / 365

plt.figure(figsize=(8,5))
plt.hist(df_attr["Age_years"], bins=60)
plt.xlabel("Age (years)")
plt.ylabel("Count")
plt.title("Distribution of patient age")
plt.show()

print("Min age (days):", df_attr["Age_days"].min())
print("Max age (days):", df_attr["Age_days"].max())
print("Average age (days):", df_attr["Age_days"].mean())

# How many have age = 0 days?
num_age_zero = (df_attr["Age_days"] == 0).sum()
print("Number of ECG records with age = 0 days:", num_age_zero)
```

```python
# ===== Gender distribution =====

plt.figure(figsize=(5,4))
df_attr["Gender"].value_counts().plot(kind="bar")
plt.title("Gender distribution")
plt.xlabel("Gender")
plt.ylabel("Count")
plt.show()

print(df_attr["Gender"].value_counts())
```

```python
# ===== Number of electrodes per ECG (9 vs 12) =====

plt.figure(figsize=(5,4))
df_attr["Lead"].value_counts().plot(kind="bar")
plt.title("Number of leads per ECG")
plt.xlabel("Lead count")
plt.ylabel("Number of ECG records")
plt.show()

print(df_attr["Lead"].value_counts())
```

```python
# Ensure Age_days exists
df_attr["Age_days"] = df_attr["Age"].str.replace("d", "").astype(int)
df_attr["Age_years"] = df_attr["Age_days"] / 365

# Split by lead type
df_9 = df_attr[df_attr["Lead"] == 9]
df_12 = df_attr[df_attr["Lead"] == 12]

print("Average age (9-lead):", df_9["Age_years"].mean())
print("Average age (12-lead):", df_12["Age_years"].mean())
print("Median age (9-lead):", df_9["Age_years"].median())
print("Median age (12-lead):", df_12["Age_years"].median())
print("Min age 9-lead:", df_9["Age_years"].min())
print("Min age 12-lead:", df_12["Age_years"].min())
print("Max age 9-lead:", df_9["Age_years"].max())
print("Max age 12-lead:", df_12["Age_years"].max())

# Plot both age distributions
plt.figure(figsize=(10,5))
plt.hist(df_9["Age_years"], bins=40, alpha=0.6, label="9-lead")
plt.hist(df_12["Age_years"], bins=40, alpha=0.6, label="12-lead")
plt.xlabel("Age (years)")
plt.ylabel("Count")
plt.title("Age distribution: 9-lead vs 12-lead ECGs")
plt.legend()
plt.show()
```

```python
# ===== ECG duration distribution =====
#Assuming sampling frequency = 500 Hz

df_attr["Duration_sec"] = df_attr["Sampling_point"] / 500

plt.figure(figsize=(8,5))
plt.hist(df_attr["Duration_sec"], bins=40)
plt.xlabel("Duration (seconds)")
plt.ylabel("Count")
plt.title("Distribution of ECG signal length")
plt.show()

print("Min duration:", df_attr["Duration_sec"].min(), "sec")
print("Max duration:", df_attr["Duration_sec"].max(), "sec")
print("Average duration:", df_attr["Duration_sec"].mean(), "sec")

# Distribution in a table
# Create bins of 5 seconds
bins = list(range(0, int(df_attr["Duration_sec"].max()) + 5, 5))

df_attr["Duration_bin"] = pd.cut(
    df_attr["Duration_sec"],
    bins=bins,
    right=False,   # intervals like [05), [510), ...
    labels=[f"{b}-{b+5}s" for b in bins[:-1]]
)

duration_table = df_attr["Duration_bin"].value_counts().sort_index()
duration_table
```

```python
# === Define cleaned cardiovascular ICD codes ===
cardio_icds = [
    "I40.0", "I40.9", "I51.4",
    "I42.0", "I42.2", "I42.9", "Q24.8",
    "M30.3",
    "Q21.0", "Q21.1", "Q21.2", "Q21.3",
    "Q22.1", "Q25.0", "Q25.6", "I37.0"
]

# === Function to clean ICD codes as in preprocessing ===
def clean_icd_str(icd):
    icd = icd.strip().replace("'", "")
    if ")" in icd:
        icd = icd.split(")")[-1].strip()
    return icd

# === Extract and clean all ICD codes ===
codes = (
    df_attr["ICD-10 code"]
    .dropna()
    .astype(str)
    .str.split(";")
)

clean_flat = []

for lst in codes:
    for x in lst:
        cleaned = clean_icd_str(x)
        clean_flat.append(cleaned)

# === Keep only cardiovascular ICDs ===
cardio_only = [c for c in clean_flat if c in cardio_icds]

# === Count ===
icd_counts_cardio = pd.Series(cardio_only).value_counts().sort_values(ascending=False)

# === Plot ===
plt.figure(figsize=(10,6))
icd_counts_cardio.plot(kind="bar")
plt.title("Distribution of Pediatric Cardiovascular ICD-10 Codes")
plt.xlabel("ICD-10 code")
plt.ylabel("Count")
plt.xticks(rotation=45)
plt.show()

icd_counts_cardio
```

```python
# === Load disease mapping from CSV ===
df_map = pd.read_csv(DISEASE_PATH, encoding='latin1')

# === Function to clean ICD codes as in preprocessing ===
def clean_icd_str(icd):
    icd = str(icd).strip().replace("'", "")
    if ")" in icd:
        icd = icd.split(")")[-1].strip()
    return icd

# Clean ICD codes in mapping file
df_map["ICD_clean"] = df_map["ICD-10 Code"].apply(clean_icd_str)

# === Create dictionary: ICD -> list of Disease Types ===
icd_to_types = {}

for _, row in df_map.iterrows():
    icd = row["ICD_clean"]
    dtype = row["Disease Type"]
    icd_to_types.setdefault(icd, []).append(dtype)

# === Load attributes dataset ===
df_attr = pd.read_csv(ATTR_PATH)

# === Function to clean ICD codes as in preprocessing ===
def clean_icd_str(icd):
    icd = str(icd).strip().replace("'", "")
    if ")" in icd:
        icd = icd.split(")")[-1].strip()
    return icd

# Extract ICD codes from dataset
codes = (
    df_attr["ICD-10 code"]
    .dropna()
    .astype(str)
    .str.split(";")
)

# Clean and flatten ICD list
clean_flat = []
for lst in codes:
    for x in lst:
        cleaned = clean_icd_str(x)
        clean_flat.append(cleaned)

# === Count disease types, allowing multiple assignments ===
type_counts = {}

for icd in clean_flat:
    if icd in icd_to_types:
        types = icd_to_types[icd]
        for t in types:
            type_counts[t] = type_counts.get(t, 0) + 1

type_counts = pd.Series(type_counts).sort_values(ascending=False)

# === Plot ===
plt.figure(figsize=(10,6))
type_counts.plot(kind="bar")
plt.title("Counts of ICD Codes Grouped by Pediatric Cardiovascular Disease Types")
plt.xlabel("Disease Type")
plt.ylabel("Count")
plt.xticks(rotation=45)
plt.show()

type_counts
```

```python
# === Load disease mapping from CSV ===
df_map = pd.read_csv(DISEASE_PATH, encoding='latin1')

# === Function to clean ICD codes as in preprocessing ===
def clean_icd_str(icd):
    icd = str(icd).strip().replace("'", "")
    if ")" in icd:
        icd = icd.split(")")[-1].strip()
    return icd

# Clean ICD codes in mapping file
df_map["ICD_clean"] = df_map["ICD-10 Code"].apply(clean_icd_str)

# === Create dictionary: ICD -> list of Disease Types ===
icd_to_types = {}
for _, row in df_map.iterrows():
    icd = row["ICD_clean"]
    dtype = row["Disease Type"]
    icd_to_types.setdefault(icd, []).append(dtype)

# === Load attributes dataset ===
df_attr = pd.read_csv(ATTR_PATH)

# === Extract ONLY the primary ICD code (first code in the line) ===
df_attr["ICD_primary"] = (
    df_attr["ICD-10 code"]
    .dropna()
    .astype(str)
    .apply(lambda s: s.split(";")[0])   # Only first ICD code
)

# Clean the primary ICD codes
df_attr["ICD_primary_clean"] = df_attr["ICD_primary"].apply(clean_icd_str)

# === Count disease types using ONLY primary ICD code ===
type_counts_primary = {}

for icd in df_attr["ICD_primary_clean"]:
    if icd in icd_to_types:
        types = icd_to_types[icd]   # Some ICDs belong to multiple disease types
        for t in types:
            type_counts_primary[t] = type_counts_primary.get(t, 0) + 1

type_counts_primary = pd.Series(type_counts_primary).sort_values(ascending=False)

# === Plot ===
plt.figure(figsize=(10,6))
type_counts_primary.plot(kind="bar")
plt.title("Counts of PRIMARY ICD Codes Grouped by Pediatric Cardiovascular Disease Types")
plt.xlabel("Disease Type")
plt.ylabel("Count")
plt.xticks(rotation=45)
plt.show()

type_counts_primary
```
